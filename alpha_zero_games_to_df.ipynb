```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("alpha0_fixed.csv",index_col=False, header=None)
```

```python
df
```

```python
df[7]=df[7].apply(lambda x : x.replace("\n"," "))

```

```python
df = df.drop([0,1,2,3], axis=1)
```

```python
df = df.rename(index=str, columns={4: 'white', 5:'black',6:'score',7:'moves'})
```

```python
df['white']=df['white'].apply(lambda x: str.split(x)[1])
df['black']=df['black'].apply(lambda x: str.split(x)[1])
```

```python
df = df.replace('"AlphaZero"]',"AlphaZero")
df = df.replace('"Stockfish',"Stockfish8")
df = df.replace('[Result "1/2-1/2"]',0.5)
df = df.replace('[Result "1-0"]',1.0)
df = df.replace('[Result "0-1"]',0.0)
```

```python
df['moves'] = df['moves'].apply(lambda x: x.split(" "))
```

```python
df['move1w'] = df['moves'].apply(lambda x: x[1])
df['move1b'] = df['moves'].apply(lambda x: x[2])
df['move2w'] = df['moves'].apply(lambda x: x[4])
df['move2b'] = df['moves'].apply(lambda x: x[5])
df['move3w'] = df['moves'].apply(lambda x: x[7])
df['move3b'] = df['moves'].apply(lambda x: x[8])
df['move4w'] = df['moves'].apply(lambda x: x[10])
df['move4b'] = df['moves'].apply(lambda x: x[11])
df['white_moves'] = df['move1w'] + df['move2w'] + df['move3w'] + df['move4w']
df['black_moves'] = df['move1b'] + df['move2b'] + df['move3b'] + df['move4b']
df['white_castle'] = df['white_moves'].str.contains('O-O',regex=False) | df['white_moves'].str.contains('O-O-O',regex=False)
df['black_castle'] = df['black_moves'].str.contains('O-O',regex=False) | df['black_moves'].str.contains('O-O-O',regex=False) 
```

```python
df.to_csv("alpha0_df.csv")
```
