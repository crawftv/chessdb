```python
df = pd.read_csv("ficsgamesdb_2017_standard_nomovetimes_34352.pgn.bz2", header=None)
```

```python
df[1]= df[0].str.contains('WhiteElo', regex=False)
df[2]=df[0].str.contains('BlackElo', regex=False)
df[3]=df[0].str.contains('Result', regex=False)
df[4]=df[0].str.contains('1. ', regex=False)
```

```python
d = df[(df[1]==True) | (df[2] ==True) | (df[3] == True) | (df[4] == True)]
len(d.values), d
```

```python
e = np.array(d[0][:4178900])
```

```python
np.reshape(e, (1044725,4 ))
```

```python
f = pd.DataFrame(np.reshape(e, (1044725,4 )))
```

```python
f.head()
```

```python
f['WhiteEloTF'] = f[0].apply(lambda row : True if "WhiteElo" in row else False)
f['BlackEloTF'] = f[1].apply(lambda row : True if "BlackElo" in row else False)
f['ResultTF'] = f[2].apply(lambda row : True if "Result" in row else False)
f['MovesTF'] = f[3].apply(lambda row : True if "1. " in row else False)
```

```python
g = f[(f['WhiteEloTF'] == True) & (f['BlackEloTF'] == True) & (f['ResultTF'] == True) & (f['MovesTF'] == True)]
```

```python
g['WhiteElo'] = g[0].str.split('"')
g['BlackElo'] = g[1].str.split('"')
```

```python
g['WhiteElo'] =g['WhiteElo'].apply(lambda x : int(x[1]) )
g['BlackElo'] =g['BlackElo'].apply(lambda x : int(x[1]) )
```

```python
g['WhiteElo'][0]
```

```python
#https://www.ficsgames.org/stats/9999_rdistr_3.html
```

```python
g[g['WhiteElo']<1450].count()
```

```python
g['moves_list'] = g[3].str.split(" ")
```

```python
g = g.rename(index=str, columns={2: "Wins"})
```

```python
g
```

```python
g['Wins'] = g['Wins'].replace('[Result "0-1"]', 0)
g['Wins'] = g['Wins'].replace('[Result "1-0"]', 1)
g['Wins'] = g['Wins'].replace('[Result "1/2-1/2"]', 0.5)
```

```python
g['move1w'] = g['moves_list'].apply(lambda x : x[1] )
g['move1b'] = g['moves_list'].apply(lambda x : x[2] if len(x)>=3 else "" )
g['move2w'] = g['moves_list'].apply(lambda x : x[4] if len(x)>=5 else "" )
g['move2b'] = g['moves_list'].apply(lambda x : x[5] if len(x)>=6 else "" )
g['move3w'] = g['moves_list'].apply(lambda x : x[7] if len(x)>=8 else "" )
g['move3b'] = g['moves_list'].apply(lambda x : x[8] if len(x)>=9 else "" )
g['move4w'] = g['moves_list'].apply(lambda x : x[10] if len(x)>=11 else "" )
g['move4b'] = g['moves_list'].apply(lambda x : x[11] if len(x)>=12 else "" )
```

```python
g['white_moves'] = g['move1w'] + g['move2w'] + g['move3w'] + g['move4w']
g['black_moves'] = g['move1b'] + g['move2b'] + g['move3b'] + g['move4b']
g['white_castle'] = g['white_moves'].str.contains('O-O',regex=False) | g['white_moves'].str.contains('O-O-O',regex=False)
g['white_checks'] = g['white_moves'].str.contains('+',regex=False)
g['black_castle'] = g['black_moves'].str.contains('O-O',regex=False) | g['black_moves'].str.contains('O-O-O',regex=False) 
g['black_checks'] = g['black_moves'].str.contains('+',regex=False)
```

```python
moves_list = ['move1w','move1b','move2w','move2b','move3w','move3b','move4w','move4b']
for i in moves_list:
    g[i] = g[i].apply(lambda x: x if '{' not in x else "")
    g[i] = g[i].apply(lambda x: x if 'drawn' not in x else "")
    g[i] = g[i].apply(lambda x: x if 'forfeits' not in x else "")
    g[i] = g[i].apply(lambda x: x if 'resigns' not in x else "")
    
```

```python
g['AverageElo'] = (g['WhiteElo'] + g['BlackElo'])/2
```

```python
g
```

```python
g['10_buckets'] = pd.cut(g['AverageElo'],10, precision =0)
```

```python
g['10_q_buckets'] = pd.qcut(g['AverageElo'],10, precision =0)
```

```python
g['5_q_buckets'] = pd.qcut(g['AverageElo'],5, precision =0)
```
