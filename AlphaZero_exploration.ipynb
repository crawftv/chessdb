```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("alpha0_df.csv",index_col=None)
df.drop(columns=['Unnamed: 0'])
```

## Score

```python
df.groupby(['white','score']).agg({'score':np.size}).unstack(level=1)
```

#### AlphaZero
- wins   = 36 = 31+5 
- losses =  3 =  1+2
- draws  = 72 = 49+23

Average score was 0.65

#### Stockfish8
- wins = 3
- losses = 36
- draws = 72

Average score was 0.35

## Castling in the First Four Moves
I noticed that castling came up more in these first four moves than my other databases. Let's explore.

```python
df.groupby(['white','white_castle','score']).agg({'white_castle':np.size}).unstack(level=2)
```

```python
print("AlphaZero was white "+str(1+49+31) +" times and did not castle once in the first four  moves")
print("Stockfish8 was white "+str(2+16+1+3+7+1) +" times and castled " + str(3+7+1)+ " times in the first four  moves. " +str(round(((3+7+1)/(2+16+1+3+7+1))*100,2)) + "%")
print("The average score for Stockfish8 when castling was " + str(round(((1+7/2)/11),2)) + ". and " +  str(round(((1+16/2)/19),2)) + " without castling" )
```

```python
df.groupby(['black','black_castle','score']).agg({'black_castle':np.size}).unstack(level=2).fillna(0)
```

```python
print("AlphaZero was black "+str(5+22+2+1) +" times and castled once in the first four moves. "+ str( round((1/(5+22+2+1))*100,2 ) ) + "%")
print("Stockfish8 was black "+str(1+38+25+11+6) +" times and castled " + str(11+6)+ " times in the first four  moves. " +str(round(((11+6)/(1+38+25+11+6))*100,2)) + "%")
print("AlphaZero had an average score of 0.5 when castling and " +  str(round(((5+22/2)/(5+22+2)),2)) + " without castling")
print("The average score for Stockfish8 when castling was " + str(round(((11/2)/17),2)) + ". and " +  str(round(((1+38/2)/(1+38+25)),2)) + " without castling" )
```

```python
print("In total Stockfish8 castled 28 times in the first four moves, winning once, for an average score of "+ str(round((10/28),2)))
```
