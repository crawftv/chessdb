### Hypothesis: Within the first four moves, Low-Elo chess players are significantly different than High-Elo players.

```python
z = b.groupby(['m1','5_q_buckets']).agg({'m1':np.size})
z = z.sort_values('m1',ascending=False).unstack(level=1).fillna(0)
chi = scipy.stats.chi2_contingency(z)
print("Test Statistic for a chi squared test is "+ str(chi[0]))
print("The p-value for a chi squared test is "+ str(chi[1]))
```

```python
z = b.groupby(['move1w','5_q_buckets']).agg({'move1w':np.size})
z = z.sort_values('move1w',ascending=False).unstack(level=1).fillna(0)
chi = scipy.stats.chi2_contingency(z)
print("Test Statistic for a chi squared test is "+ str(chi[0]))
print("The p-value for a chi squared test is "+ str(chi[1]))
```

```python
z = b[b['move1w'] == 'e4'].groupby([b['move1b'],'5_q_buckets']).agg({'move1b':np.size})
z = z.sort_values('move1b',ascending=False).unstack(level=1).fillna(0)
chi = scipy.stats.chi2_contingency(z)
print("Test Statistic for a chi squared test is "+ str(chi[0]))
print("The p-value for a chi squared test is "+ str(chi[1]))
```
